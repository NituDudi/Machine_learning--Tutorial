<h3><a href="https://www.kaggle.com/datasets/ahsan81/hotel-reservations-classification-dataset" style="text-decoration: underline;">Dataset used</a></h3>

```python
import pandas as pd

df= pd.read_csv("hotel_reservations.csv")
df.head()
```

```python
df.isna().sum()
```

```python
df['booking_status'].value_counts()
```

```python
df.drop(["booking_id", "arrival_year", "arrival_month", "arrival_date"], axis= 1,inplace= True)
```

```python
df.head()
```

```python
import seaborn as sns
from matplotlib import pyplot as plt

sns.countplot(x= "booking_status", data= df)
```

```python
sns.boxplot(y= "lead_time", data= df)
```

```python
sns.boxplot(x= "booking_status", y= "avg_price_per_room", data= df)
```

```python
df_encoded= pd.get_dummies(df, columns= ["type_of_meal_plan", "room_type_reserved", "market_segment_type"], drop_first= True)
df_encoded.head()
```

```python
from sklearn.preprocessing import StandardScaler

scale= StandardScaler()
df_encoded[['lead_time', 'avg_price_per_room']]= scale.fit_transform(df_encoded[['lead_time', 'avg_price_per_room']])

X= df_encoded.drop("booking_status", axis= 1)
y= df['booking_status']
```

```python
df_encoded.head()
```

```python
from sklearn.model_selection import cross_val_score, KFold
from sklearn.naive_bayes import GaussianNB
import numpy as np

kf= KFold(n_splits= 7, shuffle= True, random_state= 42)

nb= cross_val_score(GaussianNB(), X, y, cv= kf)
np.average(nb)
```

```python
from sklearn.linear_model import LogisticRegression

lr= cross_val_score(LogisticRegression(), X, y, cv= kf)
np.average(lr)
```

```python
from sklearn.tree import DecisionTreeClassifier

dt= cross_val_score(DecisionTreeClassifier(), X, y, cv= kf)
np.average(dt)
```

<h3>Implementing Stratified k-fold Cross Validation</h3>

```python
from sklearn.model_selection import StratifiedKFold

skf= StratifiedKFold(n_splits= 7, shuffle= True, random_state= 42)

nb= cross_val_score(GaussianNB(), X, y, cv= skf)
np.average(nb)
```

```python
lr= cross_val_score(LogisticRegression(), X, y, cv= skf)
np.average(lr)
```

```python
dt= cross_val_score(DecisionTreeClassifier(), X, y, cv= skf)
np.average(dt)
```

<h3>Implementing GridSearchCV</h3>

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier

dict = {
    'nb': {
        'model': GaussianNB(),
        'params': {'var_smoothing': [1e-9, 1e-8, 1e-7]}
    },
    'rf': {
        'model': RandomForestClassifier(),
        'params': {
            'n_estimators': [10, 50, 100],
            'max_depth': [5, 10, 15]
        }
    }
}

result= []

for key, val in dict.items():
    clf= GridSearchCV(val['model'], val['params'], cv= 5, return_train_score= False)
    clf.fit(X, y)
    result.append({
        'model': key,
        'best_params': clf.best_params_,
        'score': clf.best_score_
    })

result
    
```

```python
clf1= GridSearchCV(GaussianNB(), {'var_smoothing': [1e-9, 1e-8, 1e-7]}, cv= 5, return_train_score= False)
clf1.fit(X, y)
clf1_result= pd.DataFrame(clf1.cv_results_)


clf2 = GridSearchCV(RandomForestClassifier(), {'n_estimators': [10, 50, 100],'max_depth': [5, 10, 15]}, cv= 5, return_train_score= False)
clf2.fit(X, y)
clf2_result= pd.DataFrame(clf2.cv_results_)

clf1_result
```

```python
from sklearn.model_selection import RandomizedSearchCV

clf= RandomizedSearchCV(RandomForestClassifier(), {'n_estimators': [10, 50, 100], 'max_depth': [5, 10, 15, 20]}, cv= 5, n_iter= 3, return_train_score = False)
clf.fit(X, y)

result_clf= pd.DataFrame(clf.cv_results_)
result_clf
```
